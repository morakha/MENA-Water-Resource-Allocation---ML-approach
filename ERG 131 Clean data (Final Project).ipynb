```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df1 = pd.read_csv('aquastat_flat.csv',lineterminator='\r')
df2 = pd.read_csv('as2.csv',lineterminator='\r')[:-1]
df3 = pd.read_csv('as3.csv',lineterminator='\r')[:-1]
df4 = pd.read_csv('as4.csv',lineterminator='\r')[:-1]
df5 = pd.read_csv('as5.csv',lineterminator='\r')[:-1]
df6 = pd.read_csv('as6.csv',lineterminator='\r')[:-1]
df7 = pd.read_csv('as7.csv',lineterminator='\r')[:-1]
df8 = pd.read_csv('as8.csv',lineterminator='\r')[:-1]
df9 = pd.read_csv('as9.csv',lineterminator='\r')[:-1]
df10 = pd.read_csv('as10.csv',lineterminator='\r')[:-1]
df11 = pd.read_csv('as11.csv',lineterminator='\r')[:-1]
df12 = pd.read_csv('as12.csv',lineterminator='\r')[:-1]
df13 = pd.read_csv('as13.csv',lineterminator='\r')[:-1]
df14 = pd.read_csv('as14.csv',lineterminator='\r')[:-1]
```

```python
df = pd.concat([df1,df2,df3,df4,df5,df6,df7,df8,df9,df10,df11,df12,df13,df14])
```

```python
df.Area.replace(regex=True,inplace=True,to_replace=r'\n',value=r'')
df
```

```python
clean_df =  pd.pivot_table(df,index = ['Area','Year'],columns = 'Variable Name', values = 'Value')
clean_df = clean_df.reset_index()
```

```python
g50r = clean_df[clean_df.count(axis = 1) >= 50]
#len(g50r.Area.unique())
g50r.groupby("Area").count().sort_values(by ='Year',ascending = False).reset_index()
```

```python
countries = ['Armenia','Azerbaijan','Morocco','Algeria','Tunisia','Egypt','Turkey','Greece','Cyprus']
test = g50r[g50r.Area.isin(countries)]
#test.groupby("Year").count().mean().sort_values(ascending = False)
test1 = test[test.apply(lambda x : x.Year % 5 == 2 ,axis = 1)].groupby("Year").count().mean().sort_values(ascending = False)
variables = test1[test1 > len(countries) - 4].index.values
dummy = test[['Area','Year']]
variables
```

```python
goal = test[variables]
goal = goal.interpolate()
goal['Year'] = dummy['Year']
goal['Area'] = dummy['Area']
goal
```

```python
filter_fn1 = lambda y: y["Area"] == country and y["Year"] == x - 10
filter_fn2 = lambda y: y["Area"] == country and y["Year"] == x - 5
filter_fn3 = lambda y: y["Area"] == country and y["Year"] == x

all_rows = []
for country in countries:
    for var in variables:
        for x in range(2002, 2017, 5):
            new_row = {"Predicted Year": x, "Country": country, "Variable": var, "Year - 10": goal[goal.apply(filter_fn1, axis=1)].loc[:,var].values[0], "Year - 5": goal[goal.apply(filter_fn2, axis=1)].loc[:,var].values[0], "Predicted Value": goal[goal.apply(filter_fn3, axis=1)].loc[:,var].values[0]}
#             , "x2": goal[goal.apply(filter_fn2, axis=1)][var], "x3":goal[goal.apply(filter_fn3, axis=1)][var
            all_rows.append(new_row)
aqua_training = pd.DataFrame(all_rows)

# goal[goal.apply(lambda y: y["Area"] == "Armenia"  and y["Year"] == 2002, axis=1)].loc[:, "Rural population"].values[0]
```

```python
aqua_training = aqua_training[['Country', 'Variable', 'Predicted Year', 'Year - 10',
       'Year - 5', 'Predicted Value']]
aqua_training.set_index("Country",inplace=True)
aqua_training
# aqua_training.to_csv('aqua_training.csv')
```
